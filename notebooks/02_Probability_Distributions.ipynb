# Importing Packages

```python
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from utils import *

warnings.simplefilter(action='ignore', category=Warning)

sns.set()
```

# Loading Data

```python
# Load the iris dataset
data = load_iris()

# Get the feature matrix X and target vector y
X = data.data
y = data.target

# Get column names for features and target
feature_names = data.feature_names
target_names = data.target_names

# Combine feature matrix X and target vector y into a single DataFrame
df = pd.DataFrame(X, columns=feature_names)
df['target'] = pd.Series(y).map(dict(zip(range(len(target_names)), target_names)))

# Display the DataFrame
print(df.head())
```

# Binomial Distribution

In this case, the function generates a single random number that represents the number of successes (i.e., the number of "heads" if you were flipping a fair coin) in a sequence of 6 independent trials, where the probability of success in each trial is 0.5 (i.e., the coin has a 50/50 chance of landing "heads" or "tails" on each flip).

```python
np.random.binomial(6, 0.5)
```

In this case, the function generates 10 random numbers, where each number represents the number of successes (i.e., the number of "heads" if you were flipping a fair coin) in a sequence of 6 independent trials, where the probability of success in each trial is 0.5 (i.e., the coin has a 50/50 chance of landing "heads" or "tails" on each flip).

```python
np.random.binomial(6, 0.5, size=10)
```

## The Binomial CDF

```python
samples = np.random.binomial(60, 0.1, size=10000)

x, y = ecdf(samples)
_ = plt.plot(x, y, marker='.', linestyle='none')
plt.margins(0.02)
_ = plt.xlabel('number of successes')
_ = plt.ylabel('CDF')
plt.show()
```

# Poisson Distribution

## The Poisson CDF

```python
samples = np.random.poisson(10, size=10000)
x, y = ecdf(samples)
_ = plt.plot(x, y, marker='.', linestyle='none')
plt.margins(0.02)
_ = plt.xlabel('number of successes')
_ = plt.ylabel('CDF')
plt.show()
```

# Normal Distribution

# Cheking Normality of Iris data

```python
mean = np.mean(df['sepal length (cm)'])
std = np.std(df['sepal length (cm)'])
samples = np.random.normal(mean, std, size=10000)
x, y = ecdf(df['sepal length (cm)'])
x_theor, y_theor = ecdf(samples)

_ = plt.plot(x_theor, y_theor)
_ = plt.plot(x, y, marker='.', linestyle='none')
_ = plt.xlabel('sepal length (cm)')
_ = plt.ylabel('CDF')
plt.show()
```
